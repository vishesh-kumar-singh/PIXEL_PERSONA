```python
# splitting the dataset
X_train, X_test, y_train, y_test = train_test_split( X, y, train_size=0.89, shuffle=True, random_state=11)
```

# **DEFINING A LOGISTIC REGRESSION CLASS:**

```python
class Regression():
  def __init__(self,learning_rate,iterations):

    self.learning_rate = learning_rate
    self.iterations = iterations

  def fit(self,X,Y):

    self.m, self.n = X.shape
    self.W = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y

    for i in range(self.iterations):
      self.update_weights()

    return self

  def update_weights(self):

    A = 1 / (1+ np.exp (- ( self.X.dot( self.W ) + self.b )))
    tmp = A - self.Y.T
    tmp = np.reshape( tmp, self.m )
    dW = np.dot( self.X.T, tmp/self.m )
    db = np.sum(tmp)/self.m

    self.W = self.W-self.learning_rate *dW
    self.b = self.b-self.learning_rate *db

    return self

  def predict(self,X):

    Z = 1 / ( 1 + np.exp( -( X.dot( self.W ) + self.b )))
    Y = np.where( Z>0.5, 1, 0 )

    return Y
```

# TRAINING THE MODEL:


---


**first_model = model trained using scratch**


---


**second_model = model trained using Scikit Learn Library**


---







```python
first_model = Regression(learning_rate=0.01, iterations=1000)
first_model.fit(X_train, y_train)
second_model = LogisticRegression()
second_model.fit(X_train, y_train)
```

# CALCULATING THE ACCURACY

```python
y_pred_1 = first_model.predict(X_test)
y_pred_2 = second_model.predict(X_test)
cc_1=0
cc_2=0
```

```python
print(y_pred_1)
print(X_test)
```

```python
c=0
for c in range(np.size(y_pred_1)):

  if y_test[c] == y_pred_1[c]:
    cc_1 +=1
  if y_test[c] == y_pred_2[c]:
    cc_2 +=1
  c+=1
```

```python
print( "Accuracy on test set by first model  :  ", (cc_1 / c) * 100)
print( "Accuracy on test set by second model (sklearn model)   :  ", (cc_2 / c) * 100)
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
y=encode.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25, shuffle=True, random_state=5)
```

```python
reg=LogisticRegression()
```

```python
class LogReg:

  def __init__( self, learning_rate, iterations ) :
    self.learning_rate = learning_rate
    self.iterations = iterations

  def fit(self,X,y):
    self.m, self.n = X.shape
    self.w = np.zeros(self.n)
    self.b=0
    self.X=X
    self.y=y

    for i in range(self.iterations):
      self.updateweighs()
      return self

  def updateweighs(self):
    A = 1 / ( 1 + np.exp( - ( self.X.dot( self.w ) + self.b ) ) )
    tmp = ( A - self.y.T )
    tmp = np.reshape( tmp, self.m )
    dw = np.dot( self.X.T, tmp ) / self.m
    db = np.sum( tmp ) / self.m
    self.w = self.w - self.learning_rate * dw
    self.b = self.b - self.learning_rate * db
    return self

  def predict( self, X ) :
    Z = 1 / ( 1 + np.exp( - ( X.dot( self.w ) + self.b ) ) )
    y = np.where( Z > 0.5, 1, 0 )
    return y

model1 = LogReg( learning_rate = 0.01, iterations = 1000 )
model1.fit( X_train, y_train )
y_predict1 = model1.predict( X_test )

correctly_classified1 = 0
count1 = 0
for count1 in range( np.size( y_predict1 ) ) :
  if y_test[count1] == y_predict1[count1] :
    correctly_classified1 = correctly_classified1 + 1

acc1=((correctly_classified1)/(count1))*100
```

```python
reg.fit(X_train,y_train)
```

```python
y_predict=reg.predict(X_test)
```

```python
count=0
correctly_classified=0
for count in range( np.size( y_predict ) ) :
  if y_test[count] == y_predict[count] :
    correctly_classified = correctly_classified + 1
acc=((correctly_classified)/(count))*100
```

```python
print("The accuracy of model without sklearn library is: ",acc1,"%")
print("The accuracy of the model with sklearn is: ",acc,"%")
```
